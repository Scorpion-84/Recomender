```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('mycsv.csv')
df
```

```python
df.info()
df.shape
```

```python
plt.figure(figsize = (40,8))
plt.scatter(df['Series_Title'], df['Genre'], label='Genre')
plt.legend()
plt.xlabel('Title')
plt.ylabel('Genre')
plt.xticks(rotation = 45)
```

```python
plt.figure(figsize = (40,8))
plt.bar(df['Series_Title'], df['IMDB_Rating'], label='Rating')
plt.plot(df['Series_Title'], df['IMDB_Rating'], "r--x",label='Rating')
plt.legend()
plt.xlabel('Title')
plt.ylabel('Rating')
plt.ylim(0, 10)
plt.xticks(rotation = 45)
plt.xticks(rotation = 45)
```

```python
plt.figure(figsize = (40,8))
plt.bar(df['Genre'], df['IMDB_Rating'],label='Genre')
plt.legend()
plt.xlabel('Genre')
plt.ylabel('Rating')
plt.ylim(0,10)
plt.xticks(rotation = 45)
```
